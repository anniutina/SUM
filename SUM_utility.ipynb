set up

```python
PATH = 'requests/georequests_O(area10)_D.csv'
OTP_API = "http://localhost:8080/otp/routers/default/plan"

params = get_config('data/configs/default_SUM.json')
# params = make_paths(params) # update file paths
# VoT - value of time, param
```

```python
df = pd.read_csv(PATH)
df.head(3)
```

```python
df_reqs = df.filter(['x','y','desti_x', 'desti_y', 'treq'], axis = 1).rename(
    columns = {'x': 'origin_x', 'y': 'origin_y', 'desti_x': 'destination_x', 'desti_y': 'destination_y'})

df_reqs['treq'] = pd.to_datetime(df_reqs.treq)
df_reqs.head(3)
```

<h3>query OTP for each request</h3>

```python
df_query = df_reqs.apply(lambda row: 
                 parse_OTP_response(requests.get(OTP_API, params=make_query(row.squeeze())).json()), axis=1)
```

```python
# test
df_query[0]
```

```python
for name in list(df_query[0].keys()):
    vals = []
    for i in range(len(df_query)):
        if df_query[i]['success'] == True:
            vals.append(df_query[i][name])
        else:
            vals.append(None)
    df_reqs[name] = vals
df_reqs[:3]
```
